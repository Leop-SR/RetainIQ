# Data Understanding Findings

## General Observations

1. The shape of every table matches the dataset documentation, indicating that no records are missing during the import process.

2. Trial subscriptions have both **MRR (Monthly Recurring Revenue)** and **ARR (Annual Recurring Revenue)** equal to **0**, which is consistent with the business logic since trial users do not generate revenue.

3. All date columns were imported as strings (`object` type in Pandas) and will need to be converted to proper datetime objects before performing any temporal analysis.

## Subscription Table

5. Out of **5,000** subscriptions, **4,514** are currently active because their `end_date` is null. Only **486** subscriptions have ended.

6. An account can appear multiple times in the `subscriptions` table, meaning a customer may have multiple subscription periods due to upgrades, downgrades, renewals, or plan changes.

## Support Tickets

7. The `satisfaction_score` column contains **825 missing values** out of **2,000** support tickets. These missing values likely represent customers who did not respond to the satisfaction survey rather than missing data caused by collection errors.

## Churn Events

8. The `feedback_text` column contains **148 customer comments** out of **600** churn events. This text could be used in a future extension of the project using Natural Language Processing (NLP) techniques such as sentiment analysis or keyword extraction.

9. There are **six** churn reason categories:

   * Features
   * Support
   * Budget
   * Unknown
   * Competitor
   * Pricing

## Feature Usage

10. The dataset documentation states that RavenStack provides **40 different product features**, which matches the observed data.

11. The `feature_usage` table contains **21 duplicated values** in the `usage_id` column, even though `usage_id` is documented as the primary key. This anomaly requires further investigation before the data is used for analysis.

## Accounts

12. The `accounts` table contains **500** customers, of which **110** have churned at some point, corresponding to a churn rate of approximately **22%**.

13. The dataset includes:

* 5 industry categories
* 3 subscription plans
* 5 referral sources
* 7 countries

## Relationships

14. The relationships between tables were successfully validated:

* Every subscription belongs to an existing account.
* Every feature usage record belongs to an existing subscription.
* Every support ticket belongs to an existing account.
* Every churn event belongs to an existing account.

15. The data follows the expected business granularity:

* One account can have multiple subscriptions.
* One subscription can generate multiple feature usage events.
* One account can generate multiple support tickets.
* One account can experience one or more churn-related events.


### Usage ID duplicated problem

```python
duplicates = feature_usage[
    feature_usage["usage_id"].duplicated(keep=False)
].sort_values("usage_id")

duplicates
```

```python
duplicates["usage_id"].value_counts()
```

```python
duplicate_rows = feature_usage["usage_id"].duplicated().sum()
total_rows = len(feature_usage)

print(f"{duplicate_rows} duplicated IDs out of {total_rows} rows")
print(f"{duplicate_rows / total_rows:.4%} of the dataset")
```

The feature_usage table contains 21 duplicated primary key values. Since these rows represent distinct usage events rather than duplicated observations, the duplicated identifiers appear to be a data generation issue. In a specific module, new unique identifiers will be generated while preserving the original raw dataset.
